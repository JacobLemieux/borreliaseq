```python
import pandas as pd
from Bio import SeqIO
```

```python
metadata = pd.read_csv("../results/LymeSeq_SampleTrack - Renaming Scheme Table for Jupyter.2023-01-14.csv")
metadata = metadata.dropna(subset=["Rename_A"])
```

```python
metadata
```

```python
if True:
    f = open("../results/coverage_db.txt", "w")
    for index, samplename in enumerate(metadata["Rename_A"]):
        print(index, samplename)
        for seqrecord in SeqIO.parse("../results/assemblies/" + samplename + "_200.fa", "fasta"):
            if int(seqrecord.id.split("_")[3]) > 15000:
                print(samplename.strip()+"\t"+"\t".join([seqrecord.id.strip().split("_")[i] for i in [1,3,5]]), file=f)
    f.close()           
```
